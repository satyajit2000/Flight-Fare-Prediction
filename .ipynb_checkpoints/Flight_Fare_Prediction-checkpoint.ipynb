```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel(r"Data_Train.xlsx")
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data["journey_day"] = pd.to_datetime(data["Date_of_Journey"],format = "%d/%m/%Y").dt.day
data["journey_month"] = pd.to_datetime(data["Date_of_Journey"],format = "%d/%m/%Y").dt.month
data
```

```python
data.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
data["dep_hour"] = pd.to_datetime(data.Dep_Time).dt.hour
data["dep_minute"] = pd.to_datetime(data.Dep_Time).dt.minute
data
```

```python
data.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
data
```

```python
data["arriv_hour"] = pd.to_datetime(data.Arrival_Time).dt.hour
data["arriv_min"] = pd.to_datetime(data.Arrival_Time).dt.minute
data
```

```python
data.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
data.Total_Stops.unique()
```

```python
data.replace({"non-stop":0, "1 stop":1, "2 stops":2, "3 stops":3, "4 stops":4}, inplace = True)
```

```python
duration = list(data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) !=2:
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"
        else:
            duration[i] = "0h " + duration[i]
            
duration_hour = []
duration_min = []
for i in range(len(duration)):
    duration_hour.append(int(duration[i].split(sep = "h")[0]))
    duration_min.append(int(duration[i].split(sep = "m")[0].split()[-1]))
```

```python
data["duration_hour"] = duration_hour
data["duration_minutes"] = duration_min
```

```python
data
```

```python
data.drop(["Route"], axis = 1, inplace = True)
```

```python
data.drop(["Duration","Additional_Info"], axis = 1, inplace = True)
```

```python
data
```

```python
airline = data[["Airline"]]
airline = pd.get_dummies(airline,drop_first= True)
airline.head()
```

```python
source = data[["Source"]]
source = pd.get_dummies(source,drop_first= True)
source.head()
```

```python
destination = data[["Destination"]]
destination = pd.get_dummies(destination,drop_first= True)
destination.head()
```

```python
data = pd.concat([data,airline,source,destination],axis = 1)
data.head()
```

```python
data.drop(["Airline","Source","Destination"], axis=1, inplace = True)
```
